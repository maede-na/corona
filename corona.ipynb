```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv(r'C:\Users\DELL 5110\WHO-COVID-19-global-data.csv', sep=';')

# Strip whitespace from column names
df.columns = df.columns.str.strip()

# Convert the 'Date_reported' column to datetime format
df['Date_reported'] = pd.to_datetime(df['Date_reported'], format='%d/%m/%Y')

# Group by WHO region and sum the new cases and deaths
region_summary = df.groupby('WHO_region').agg({
    'New_cases': 'sum',
    'New_deaths': 'sum'
}).reset_index()

# Sort the regions by total new cases
region_summary = region_summary.sort_values(by='New_cases', ascending=False)

# Plotting
plt.figure(figsize=(12, 6))

# Bar chart for new cases by WHO region
plt.bar(region_summary['WHO_region'], region_summary['New_cases'], color='skyblue', label='New Cases')
plt.bar(region_summary['WHO_region'], region_summary['New_deaths'], color='salmon', label='New Deaths', alpha=0.7)

# Adding titles and labels
plt.title('Total New COVID-19 Cases and Deaths by WHO Region')
plt.xlabel('WHO Region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```
